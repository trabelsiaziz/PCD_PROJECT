```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
}

# Function to download a video and save it under the appropriate directory
def download_video(video_url, save_path):
    try:
        # Send a GET request to download the video
        response = requests.get(video_url, headers=headers, stream=True)
        response.raise_for_status()  # Check if the request was successful
        
        # Open the save path and write the video content
        with open(save_path, 'wb') as file:
            for chunk in response.iter_content(chunk_size=1024):
                if chunk:
                    file.write(chunk)
        print(f"Downloaded: {save_path}")
    except Exception as e:
        print(f"Failed to download {video_url}: {e}")
```

```python
from urllib.parse import urlparse

def is_youtube_url(url):
    parsed = urlparse(url)
    return 'youtube.com' in parsed.netloc or 'youtu.be' in parsed.netloc

```

```python
def download_youtube_video(url, output_path):
    ydl_opts = {
        'format': 'bestvideo[ext=mp4]/best',  # only video (prefer mp4)
        'outtmpl': output_path,
        'quiet': True
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        try:
            ydl.download([url])
            print(f"Downloaded (video only): {output_path}")
        except Exception as e:
            print(f"Failed to download {url}: {e}")
```

```python


# Load the JSON data from the file
with open('WLASL_v0.3.json', 'r') as f:
    data = json.load(f)



# Function to organize and download videos
def organize_and_download_videos(data, root_dir="DATA_COLLECTION_version_3"):
    os.makedirs(root_dir, exist_ok=True)  
    
    for item in data:
        gloss = item['gloss']
        gloss_path = os.path.join(root_dir, gloss)
        os.makedirs(gloss_path, exist_ok=True)

        for instance in item['instances']:
            video_url = instance['url']
            video_id = instance['video_id']

            # ✅ Use video_id as the filename instead of extracting from URL
            file_name = f"{video_id}.mp4"

            video_save_path = os.path.join(gloss_path, file_name)

            if is_youtube_url(video_url) : 
                download_youtube_video(video_url, video_save_path)
            else : 
                download_video(video_url, video_save_path)


# Organize and download videos
organize_and_download_videos(data)
```
